```python
"""
Gradient descent demo.
"""

import numpy as np
```

```python
# create synthetic data
x = np.linspace(-1,1,100)
y = 2*x + np.random.randn(len(x))*0.33

%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(x,y)
```

```python
def optimize(x,y,n_iter,theta_init,lr=0.2):
    """
    Gradient descent to do least-squares line-fitting to 
    data samples in x and y.
    
    Model: y = a*x+b
    """

    # store mean-square-error from each iteration
    err = np.zeros((n_iter,))
    
    # init the parameters (a,b)
    theta_old = theta_init
    
    # create space for the gradient
    J = np.zeros((2,))
    
    # run n_iter times
    for i in np.arange(n_iter):
        
        # compute current mean-squared-error (i.e., using the 
        # current parameter estimates)
        err[i] = np.mean(np.power(y-theta_old[0]-theta_old[1]*x,2.0))
        
        # compute the gradient
        J[0] = np.mean(2*(y-theta_old[0]-theta_old[1]*x)*(-1))
        J[1] = np.mean(2*(y-theta_old[0]-theta_old[1]*x)*(-x))
        
        # gradient descent update rule with learning rate lr
        theta_new = theta_old - lr*J
        theta_old = theta_new
    
    return err, theta_old

# call gradient descent optimizer with initial parameters (a,b)=(1,1)
err,param = optimize(x,y,100,np.asarray([1,1]))

# plot the MSE over the iterations
plt.plot(err)
plt.title('Param=' + str(param))
```
