```python
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.enc = nn.Sequential(
            nn.Linear(10,5, bias=True),
            nn.ReLU(),
            nn.Linear(5,2, bias=True))
        
        self.dec = nn.Sequential(
            nn.Linear(2,5, bias=True),
            nn.ReLU(),
            nn.Linear(5,10, bias=True))
        
    def forward(self, x):
        z = self.enc(x) # encoding
        y = self.dec(z) # decoding
        return z, y
            
m = MyModel()
x = torch.randn(10)
```
